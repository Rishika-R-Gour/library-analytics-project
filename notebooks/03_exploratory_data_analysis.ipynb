## üë• **Enhanced Member Analytics & Personalization Insights**

### Advanced member behavior analysis using our recommendation engine foundation

```python
# Enhanced Member Preferences Analysis
print("üë• **ENHANCED MEMBER ANALYTICS**")
print("=" * 60)

# First check Member table structure
member_columns = pd.read_sql_query("PRAGMA table_info(Member)", conn)
print("üìã Member table columns:", member_columns['name'].tolist())

# Load member data with preferences using correct column names
member_prefs_query = """
    SELECT m.Member_ID, m.Member_Type,
           mp.Reading_Level, mp.Preferred_Format, mp.Content_Sensitivity,
           mp.Privacy_Level, mp.Email_Notifications, mp.Marketing_Opt_In,
           mp.Recommendation_Algorithm_Preference
    FROM Member m
    JOIN Member_Preferences mp ON m.Member_ID = mp.Member_ID
"""

member_prefs = pd.read_sql_query(member_prefs_query, conn)
print(f"üìä Analyzing {len(member_prefs)} members with preference profiles")

# Member Type Distribution with Preferences
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# 1. Member Type vs Reading Level
member_reading = pd.crosstab(member_prefs['Member_Type'], member_prefs['Reading_Level'])
member_reading.plot(kind='bar', ax=axes[0,0], color=['lightblue', 'lightgreen', 'lightcoral', 'lightyellow'])
axes[0,0].set_title('üìö Reading Level by Member Type', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Member Type')
axes[0,0].set_ylabel('Count')
axes[0,0].legend(title='Reading Level', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[0,0].tick_params(axis='x', rotation=45)

# 2. Preferred Format Distribution
format_dist = member_prefs['Preferred_Format'].value_counts()
axes[0,1].pie(format_dist.values, labels=format_dist.index, autopct='%1.1f%%', startangle=90)
axes[0,1].set_title('üì± Preferred Format Distribution', fontsize=14, fontweight='bold')

# 3. Communication Preferences
comm_prefs = member_prefs[['Email_Notifications', 'Marketing_Opt_In']].sum()
comm_prefs.plot(kind='bar', ax=axes[1,0], color=['skyblue', 'lightcoral'])
axes[1,0].set_title('üìß Communication Preferences', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Preference Type')
axes[1,0].set_ylabel('Members Opted In')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Recommendation Algorithm Preferences
rec_prefs = member_prefs['Recommendation_Algorithm_Preference'].value_counts()
axes[1,1].bar(rec_prefs.index, rec_prefs.values, color=['lightgreen', 'lightblue', 'lightcoral', 'lightyellow'])
axes[1,1].set_title('üéØ Recommendation Algorithm Preferences', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Algorithm Type')
axes[1,1].set_ylabel('Count')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Detailed member segmentation analysis
print(f"\nüéØ **MEMBER SEGMENTATION INSIGHTS**:")
print(f"   üìä Member Types: {member_prefs['Member_Type'].value_counts().to_dict()}")
print(f"   üìö Reading Levels: {member_prefs['Reading_Level'].value_counts().to_dict()}")
print(f"   üì± Format Preferences: {member_prefs['Preferred_Format'].value_counts().to_dict()}")

# Communication engagement analysis
email_rate = member_prefs['Email_Notifications'].mean()
marketing_rate = member_prefs['Marketing_Opt_In'].mean()
print(f"\nüìß **COMMUNICATION ENGAGEMENT**:")
print(f"   ‚úâÔ∏è Email Notifications: {email_rate:.1%} opted in")
print(f"   üéØ Marketing Communications: {marketing_rate:.1%} opted in")

# Privacy analysis
privacy_dist = member_prefs['Privacy_Level'].value_counts()
print(f"\nüîí **PRIVACY PREFERENCES**:")
for level, count in privacy_dist.items():
    print(f"   {level}: {count} members ({count/len(member_prefs):.1%})")
```

## üìà **Demand Forecasting & Inventory Optimization**

### Advanced demand analysis using reservation patterns and member engagement data

```python
# Demand Forecasting Analysis
print("üìà **DEMAND FORECASTING & INVENTORY OPTIMIZATION**")
print("=" * 70)

# Analyze reservation patterns for demand insights
reservation_analysis = pd.read_sql_query("""
    SELECT ir.Status, ir.Priority_Level,
           COUNT(*) as reservation_count,
           AVG(JULIANDAY(ir.Expected_Available_Date) - JULIANDAY(ir.Request_Date)) as avg_wait_days
    FROM Item_Reservations ir
    GROUP BY ir.Status, ir.Priority_Level
    ORDER BY ir.Status, ir.Priority_Level
""", conn)

print(f"üìã **RESERVATION PATTERNS ANALYSIS**:")
print(reservation_analysis.to_string(index=False))

# High-demand items analysis
high_demand_items = pd.read_sql_query("""
    SELECT i.Title, COUNT(ir.Reservation_ID) as total_reservations,
           COUNT(CASE WHEN ir.Status = 'Active' THEN 1 END) as active_reservations,
           COUNT(CASE WHEN ir.Status = 'Fulfilled' THEN 1 END) as fulfilled_reservations,
           AVG(ir.Priority_Level) as avg_priority
    FROM Item_Reservations ir
    JOIN Item i ON ir.Item_ID = i.Item_ID
    GROUP BY i.Item_ID, i.Title
    HAVING COUNT(ir.Reservation_ID) >= 5
    ORDER BY total_reservations DESC
    LIMIT 10
""", conn)

print(f"\nüî• **TOP 10 HIGH-DEMAND ITEMS**:")
for _, row in high_demand_items.iterrows():
    title = row['Title'][:40] + "..." if len(row['Title']) > 40 else row['Title']
    print(f"   üìö {title}")
    print(f"      üìä Total: {int(row['total_reservations'])}, Active: {int(row['active_reservations'])}, Fulfilled: {int(row['fulfilled_reservations'])}")
    print(f"      üéØ Avg Priority: {row['avg_priority']:.1f}")

# Create demand forecasting visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Reservation Status Distribution
status_counts = reservation_analysis.groupby('Status')['reservation_count'].sum()
axes[0,0].pie(status_counts.values, labels=status_counts.index, autopct='%1.1f%%', startangle=90)
axes[0,0].set_title('üìã Reservation Status Distribution', fontsize=14, fontweight='bold')

# 2. Priority Level Distribution
priority_dist = pd.read_sql_query("SELECT Priority_Level, COUNT(*) as count FROM Item_Reservations GROUP BY Priority_Level", conn)
axes[0,1].bar(priority_dist['Priority_Level'], priority_dist['count'], color=['green', 'orange', 'red'])
axes[0,1].set_title('üéØ Priority Level Distribution', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Priority Level')
axes[0,1].set_ylabel('Count')

# 3. Reservation Timeline Analysis
timeline_data = pd.read_sql_query("""
    SELECT DATE(Request_Date) as request_date, COUNT(*) as daily_requests
    FROM Item_Reservations
    GROUP BY DATE(Request_Date)
    ORDER BY request_date
""", conn)

timeline_data['request_date'] = pd.to_datetime(timeline_data['request_date'])
axes[1,0].plot(timeline_data['request_date'], timeline_data['daily_requests'], marker='o', linewidth=2)
axes[1,0].set_title('üìÖ Daily Reservation Requests Timeline', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Date')
axes[1,0].set_ylabel('Daily Requests')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Fulfillment Rate Analysis
fulfillment_data = pd.read_sql_query("""
    SELECT Library_ID,
           COUNT(*) as total_reservations,
           COUNT(CASE WHEN Status = 'Fulfilled' THEN 1 END) as fulfilled,
           ROUND(COUNT(CASE WHEN Status = 'Fulfilled' THEN 1 END) * 100.0 / COUNT(*), 1) as fulfillment_rate
    FROM Item_Reservations
    GROUP BY Library_ID
    ORDER BY Library_ID
""", conn)

axes[1,1].bar(fulfillment_data['Library_ID'], fulfillment_data['fulfillment_rate'], color='lightblue')
axes[1,1].set_title('üìä Fulfillment Rate by Library', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Library ID')
axes[1,1].set_ylabel('Fulfillment Rate (%)')
axes[1,1].set_ylim(0, 100)

plt.tight_layout()
plt.show()

# Advanced demand insights
print(f"\nüìä **DEMAND FORECASTING INSIGHTS**:")
total_reservations = pd.read_sql_query("SELECT COUNT(*) as count FROM Item_Reservations", conn).iloc[0]['count']
active_reservations = pd.read_sql_query("SELECT COUNT(*) as count FROM Item_Reservations WHERE Status = 'Active'", conn).iloc[0]['count']
avg_wait_time = pd.read_sql_query("SELECT AVG(JULIANDAY(Expected_Available_Date) - JULIANDAY(Request_Date)) as avg_wait FROM Item_Reservations WHERE Status = 'Active'", conn).iloc[0]['avg_wait']

print(f"   üìã Total Reservations: {total_reservations:,}")
print(f"   üîÑ Currently Active: {active_reservations:,}")
print(f"   ‚è±Ô∏è Average Wait Time: {avg_wait_time:.1f} days")
print(f"   üìà Overall Fulfillment Rate: {fulfillment_data['fulfillment_rate'].mean():.1f}%")
```

## ‚≠ê **Social Engagement & Community Analytics**

### Member review patterns and community engagement insights for recommendation algorithms

```python
# Social Engagement & Review Analysis
print("‚≠ê **SOCIAL ENGAGEMENT & COMMUNITY ANALYTICS**")
print("=" * 70)

# Review engagement analysis
review_stats = pd.read_sql_query("""
    SELECT Rating, Reading_Status, Would_Recommend,
           COUNT(*) as review_count,
           AVG(Helpful_Votes) as avg_helpful_votes,
           AVG(Total_Votes) as avg_total_votes,
           AVG(Difficulty_Level) as avg_difficulty
    FROM Item_Reviews
    GROUP BY Rating, Reading_Status, Would_Recommend
    ORDER BY Rating DESC, Reading_Status
""", conn)

print(f"üìä **REVIEW ENGAGEMENT OVERVIEW**:")
print(f"   ‚≠ê Total Reviews: {pd.read_sql_query('SELECT COUNT(*) as count FROM Item_Reviews', conn).iloc[0]['count']:,}")
print(f"   üëç Recommendation Rate: {pd.read_sql_query('SELECT AVG(Would_Recommend) as rate FROM Item_Reviews', conn).iloc[0]['rate']:.1%}")
print(f"   ‚úÖ Verified Borrowers: {pd.read_sql_query('SELECT AVG(Verified_Borrower) as rate FROM Item_Reviews', conn).iloc[0]['rate']:.1%}")

# Top-rated items analysis
top_rated_items = pd.read_sql_query("""
    SELECT i.Title, 
           COUNT(ir.Review_ID) as review_count,
           AVG(ir.Rating) as avg_rating,
           AVG(ir.Helpful_Votes) as avg_helpful_votes,
           COUNT(CASE WHEN ir.Would_Recommend = 1 THEN 1 END) as recommendations
    FROM Item_Reviews ir
    JOIN Item i ON ir.Item_ID = i.Item_ID
    GROUP BY i.Item_ID, i.Title
    HAVING COUNT(ir.Review_ID) >= 3
    ORDER BY avg_rating DESC, review_count DESC
    LIMIT 8
""", conn)

print(f"\nüåü **TOP RATED ITEMS** (3+ reviews):")
for _, row in top_rated_items.iterrows():
    title = row['Title'][:35] + "..." if len(row['Title']) > 35 else row['Title']
    print(f"   üìö {title}")
    print(f"      ‚≠ê {row['avg_rating']:.1f}/5.0 ({int(row['review_count'])} reviews, {int(row['recommendations'])} recommendations)")

# Create social engagement visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Rating Distribution
rating_dist = pd.read_sql_query("SELECT Rating, COUNT(*) as count FROM Item_Reviews GROUP BY Rating ORDER BY Rating", conn)
axes[0,0].bar(rating_dist['Rating'], rating_dist['count'], color=['red', 'orange', 'yellow', 'lightgreen', 'green'])
axes[0,0].set_title('‚≠ê Rating Distribution', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Rating')
axes[0,0].set_ylabel('Count')

# 2. Reading Status vs Rating
status_rating = pd.read_sql_query("""
    SELECT Reading_Status, AVG(Rating) as avg_rating, COUNT(*) as count
    FROM Item_Reviews 
    GROUP BY Reading_Status
""", conn)
axes[0,1].bar(status_rating['Reading_Status'], status_rating['avg_rating'], color=['lightblue', 'lightgreen', 'lightcoral'])
axes[0,1].set_title('üìñ Average Rating by Reading Status', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Reading Status')
axes[0,1].set_ylabel('Average Rating')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Engagement vs Rating
engagement_data = pd.read_sql_query("""
    SELECT Rating, AVG(Helpful_Votes) as avg_helpful, AVG(Total_Votes) as avg_total
    FROM Item_Reviews
    GROUP BY Rating
    ORDER BY Rating
""", conn)
axes[1,0].plot(engagement_data['Rating'], engagement_data['avg_helpful'], marker='o', label='Helpful Votes', linewidth=2)
axes[1,0].plot(engagement_data['Rating'], engagement_data['avg_total'], marker='s', label='Total Votes', linewidth=2)
axes[1,0].set_title('üëç Engagement by Rating', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Rating')
axes[1,0].set_ylabel('Average Votes')
axes[1,0].legend()

# 4. Age Appropriateness Distribution
age_dist = pd.read_sql_query("SELECT Age_Appropriateness_Rating, COUNT(*) as count FROM Item_Reviews GROUP BY Age_Appropriateness_Rating", conn)
axes[1,1].pie(age_dist['count'], labels=age_dist['Age_Appropriateness_Rating'], autopct='%1.1f%%', startangle=90)
axes[1,1].set_title('üë∂ Age Appropriateness Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# Community engagement insights
high_engagement_reviews = pd.read_sql_query("""
    SELECT COUNT(*) as count 
    FROM Item_Reviews 
    WHERE Total_Votes >= 10
""", conn).iloc[0]['count']

content_variety = pd.read_sql_query("""
    SELECT Difficulty_Level, COUNT(*) as count,
           AVG(Rating) as avg_rating
    FROM Item_Reviews
    GROUP BY Difficulty_Level
    ORDER BY Difficulty_Level
""", conn)

print(f"\nüéØ **COMMUNITY ENGAGEMENT INSIGHTS**:")
print(f"   üèÜ High Engagement Reviews (10+ votes): {high_engagement_reviews}")
print(f"   üìä Content Difficulty Distribution:")
for _, row in content_variety.iterrows():
    print(f"      Level {int(row['Difficulty_Level'])}: {int(row['count'])} reviews (Avg: {row['avg_rating']:.1f}‚≠ê)")

# Recommendation engine readiness
recommendation_data = pd.read_sql_query("""
    SELECT 
        COUNT(DISTINCT ir.Member_ID) as reviewing_members,
        COUNT(DISTINCT ir.Item_ID) as reviewed_items,
        AVG(ir.Rating) as avg_rating,
        AVG(CASE WHEN ir.Would_Recommend = 1 THEN 1.0 ELSE 0.0 END) as recommendation_rate
    FROM Item_Reviews ir
""", conn)

print(f"\nü§ñ **RECOMMENDATION ENGINE READINESS**:")
print(f"   üë• Active Reviewers: {int(recommendation_data.iloc[0]['reviewing_members'])} members")
print(f"   üìö Reviewed Items: {int(recommendation_data.iloc[0]['reviewed_items'])} items")
print(f"   ‚≠ê Overall Avg Rating: {recommendation_data.iloc[0]['avg_rating']:.1f}/5.0")
print(f"   üëç Would Recommend Rate: {recommendation_data.iloc[0]['recommendation_rate']:.1%}")
```

## üìä **Operational Intelligence & Business Performance**

### Executive dashboard insights and multi-branch performance analytics

```python
# Operational Intelligence Analysis
print("üìä **OPERATIONAL INTELLIGENCE & BUSINESS PERFORMANCE**")
print("=" * 80)

# Load and analyze daily operations data
operations_summary = pd.read_sql_query("""
    SELECT Library_ID, 
           AVG(New_Loans) as avg_daily_loans,
           AVG(Member_Visits) as avg_daily_visits,
           AVG(Penalties_Collected + Membership_Fees_Collected) as avg_daily_revenue,
           AVG(Member_Satisfaction_Score) as avg_satisfaction,
           AVG(Cost_Per_Transaction) as avg_cost_per_transaction,
           COUNT(*) as days_recorded
    FROM Daily_Operations_Summary
    GROUP BY Library_ID
    ORDER BY Library_ID
""", conn)

print("üèõÔ∏è **BRANCH PERFORMANCE OVERVIEW**:")
print(operations_summary.round(2).to_string(index=False))

# Time series analysis
time_series_data = pd.read_sql_query("""
    SELECT Summary_Date,
           SUM(New_Loans) as total_loans,
           SUM(Member_Visits) as total_visits,
           SUM(Penalties_Collected + Membership_Fees_Collected) as total_revenue,
           AVG(Member_Satisfaction_Score) as avg_satisfaction
    FROM Daily_Operations_Summary
    GROUP BY Summary_Date
    ORDER BY Summary_Date
""", conn)

time_series_data['Summary_Date'] = pd.to_datetime(time_series_data['Summary_Date'])

# Create comprehensive operational dashboard
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Branch Performance Comparison
branch_metrics = ['avg_daily_loans', 'avg_daily_visits', 'avg_daily_revenue', 'avg_satisfaction']
x = np.arange(len(operations_summary))
width = 0.2

for i, metric in enumerate(branch_metrics):
    axes[0,0].bar(x + i*width, operations_summary[metric], width, 
                  label=metric.replace('avg_', '').replace('_', ' ').title())

axes[0,0].set_title('üèÜ Branch Performance Metrics', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Library ID')
axes[0,0].set_xticks(x + width * 1.5)
axes[0,0].set_xticklabels(operations_summary['Library_ID'])
axes[0,0].legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# 2. Revenue Trends Over Time
axes[0,1].plot(time_series_data['Summary_Date'], time_series_data['total_revenue'], 
               marker='o', linewidth=2, color='green')
axes[0,1].set_title('üí∞ Daily Revenue Trends', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Date')
axes[0,1].set_ylabel('Revenue ($)')
axes[0,1].tick_params(axis='x', rotation=45)

# 3. Loans vs Visits Correlation
axes[1,0].scatter(time_series_data['total_visits'], time_series_data['total_loans'], 
                  alpha=0.6, color='blue')
axes[1,0].set_title('üìö Loans vs Visits Correlation', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Daily Visits')
axes[1,0].set_ylabel('Daily Loans')

# Add correlation line
correlation = np.corrcoef(time_series_data['total_visits'], time_series_data['total_loans'])[0,1]
z = np.polyfit(time_series_data['total_visits'], time_series_data['total_loans'], 1)
p = np.poly1d(z)
axes[1,0].plot(time_series_data['total_visits'], p(time_series_data['total_visits']), 
               "r--", alpha=0.8)
axes[1,0].text(0.05, 0.95, f'Correlation: {correlation:.3f}', transform=axes[1,0].transAxes, 
               bbox=dict(boxstyle="round", facecolor='wheat', alpha=0.5))

# 4. Member Satisfaction Trends
axes[1,1].plot(time_series_data['Summary_Date'], time_series_data['avg_satisfaction'], 
               marker='s', linewidth=2, color='orange')
axes[1,1].set_title('üòä Member Satisfaction Trends', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Date')
axes[1,1].set_ylabel('Satisfaction Score')
axes[1,1].set_ylim(3.5, 5.0)
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Key Performance Indicators
total_operations = pd.read_sql_query("""
    SELECT 
        SUM(New_Loans) as total_loans,
        SUM(Returns_Processed) as total_returns,
        SUM(Member_Visits) as total_visits,
        SUM(Penalties_Collected + Membership_Fees_Collected) as total_revenue,
        SUM(New_Registrations) as new_members,
        AVG(Member_Satisfaction_Score) as avg_satisfaction,
        COUNT(DISTINCT Summary_Date) as operating_days
    FROM Daily_Operations_Summary
""", conn)

kpis = total_operations.iloc[0]
print(f"\nüìà **KEY PERFORMANCE INDICATORS** (90-Day Period):")
print(f"   üìö Total Loans: {int(kpis['total_loans']):,}")
print(f"   üìñ Total Returns: {int(kpis['total_returns']):,}")
print(f"   üë• Total Visits: {int(kpis['total_visits']):,}")
print(f"   üí∞ Total Revenue: ${kpis['total_revenue']:,.2f}")
print(f"   üÜï New Members: {int(kpis['new_members'])}")
print(f"   üòä Avg Satisfaction: {kpis['avg_satisfaction']:.1f}/5.0")
print(f"   üìÖ Operating Days: {int(kpis['operating_days'])}")

# Efficiency metrics
print(f"\n‚ö° **EFFICIENCY METRICS**:")
print(f"   üìä Loans per Day: {kpis['total_loans']/kpis['operating_days']:.1f}")
print(f"   üë• Visits per Day: {kpis['total_visits']/kpis['operating_days']:.1f}")
print(f"   üíµ Revenue per Day: ${kpis['total_revenue']/kpis['operating_days']:.2f}")
print(f"   üìö Loans per Visit: {kpis['total_loans']/kpis['total_visits']:.2f}")

# Branch ranking
best_performers = operations_summary.nlargest(3, 'avg_daily_loans')
print(f"\nüèÜ **TOP PERFORMING BRANCHES** (by daily loans):")
for i, (_, branch) in enumerate(best_performers.iterrows(), 1):
    print(f"   #{i} Library {int(branch['Library_ID'])}: {branch['avg_daily_loans']:.1f} loans/day, "
          f"{branch['avg_satisfaction']:.1f}‚≠ê satisfaction")

conn.close()
```
